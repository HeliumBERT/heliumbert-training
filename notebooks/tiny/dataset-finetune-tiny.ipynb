```python
import os
```

```python
BASE_PATH = "../../data/finetune/tiny/QS-OCR-small"
```

```python
dataset_raw = []

for folder_name in os.listdir(BASE_PATH):
    folder_path = os.path.join(BASE_PATH, folder_name)
    for file_name in os.listdir(folder_path):
        with open(os.path.join(folder_path, file_name), mode="r", encoding="UTF-8") as f:
            dataset_raw.append({
                "text": "\n".join(f.readlines()),
                "label": folder_name.lower()
            })
```

```python
dataset_raw
```

```python
from datasets import Dataset
```

```python
dataset = Dataset.from_list(dataset_raw)
dataset
```

```python
dataset.save_to_disk(BASE_PATH + "/../QS-OCR-small-raw")
```
